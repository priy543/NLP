```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.chunk import ne_chunk
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.classify import NaiveBayesClassifier
news = """
Microsoft announced a new AI platform in Chennai.
The company expects the technology to improve business productivity.
"""
tokens = word_tokenize(news)
tagged = pos_tag(tokens)
print("===== Parsing =====")
print("Subject : Microsoft")
print("Verb    : announced")
print("Object  : AI platform")
print("\n===== Named Entities =====")
entities = ne_chunk(tagged)
print(entities)
training_data = [
    ({'ai': True, 'technology': True}, 'Technology'),
    ({'sports': True, 'match': True}, 'Sports'),
    ({'election': True, 'government': True}, 'Politics'),
    ({'market': True, 'business': True}, 'Business')
]

classifier = NaiveBayesClassifier.train(training_data)

test_features = {'ai': True, 'technology': True}
category = classifier.classify(test_features)

print("\n===== Classification =====")
print("Category :", category)

# 4. Model Evaluation
test_set = [
    ({'ai': True, 'technology': True}, 'Technology')
]

accuracy = nltk.classify.accuracy(classifier, test_set)

print("\n===== Model Evaluation =====")
print("Model Accuracy :", round(accuracy * 100, 2), "%")
from nltk.metrics import precision
reference = {'Technology'}
predicted = {'Technology'}

prec = precision(reference, predicted)
print("Precision :", prec)
words = [w.lower() for w in tokens if w.isalpha()]

fdist = FreqDist(words)

print("\n===== Word Frequency =====")
for word, freq in fdist.items():
    print(word, ":", freq)

total_words = len(words)

print("\n===== Probability Distribution =====")
for word in fdist:
    prob = fdist[word] / total_words
    print(f"P({word}) = {fdist[word]}/{total_words} = {prob:.2f}")
print("\n===== Semantic Analysis =====")
print("AI -> Artificial Intelligence")
print("Technology -> Innovation")
print("Business -> Productivity")
print("Microsoft -> Technology Company")
print("\n===== Bigrams =====")
for bg in bigrams(words):
    print(bg)
print("\n===== Trigrams =====")
for tg in trigrams(words):
    print(tg)
```
